```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi, linspace, sin
```

```python
X = [i for i in range(22,32)]
Y = [150,155,160,162,171,174,180,183,189,192]

# fig, ax = plt.plot()
plt.plot(X,Y,marker='o')
plt.grid()
```

```python
import random
k = random.uniform(-5,5)
c = random.uniform(-5,5)

Yp = [0,0]
Yp[0] = k*X[0]+c
Yp[1] = k*X[-1]+c
plt.plot(X,Y,marker='o')
plt.plot([X[0],X[-1]],Yp,marker='o')
plt.grid()
print('Y =', k, '*X + ', c)
```

```python
data = {X[i]:Y[i] for i in range(len(X))}
```

```python
def proceed(x):
    return x*k+c
```

```python
rate = 0.00001
for i in range(100000):
    x = np.array(random.choice(list(data.keys())))
    true_result = data[x.tolist()]
    out = proceed(x)
    delta = true_result - out
    k += delta*rate*x
    c += delta*rate

Yp = [proceed(X[0]),proceed(X[-1])]
plt.plot(X,Y,marker='o')
plt.plot([X[0],X[-1]],Yp,marker='o')
plt.grid()
print('Y =', k, '*X + ', c)
```
