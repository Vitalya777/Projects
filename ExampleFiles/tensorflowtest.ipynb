```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('training data set', X_train.shape, 'labels of the training dataset', y_train.shape)
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
plt.colorbar()
plt.grid()
plt.show()
```

```python
n = 4
plt.figure(figsize=(10, 10))
for i in range(n**2):
    plt.subplot(n, n, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(X_train[i], cmap=plt.cm.binary)
plt.show()
```
