```python
from imageai.Classification import ImageClassification
import os
import torch
```

```python
execution_path = os.getcwd()

prediction = ImageClassification()
```

```python
prediction.setModelTypeAsMobileNetV2()
prediction.setModelPath(os.path.join(execution_path, "Models/mobilenet_v2-b0353104.pth"))
```

```python
prediction.setModelTypeAsInceptionV3()
prediction.setModelPath(os.path.join(execution_path, "Models/inception_v3_google-1a9a5a14.pth"))
```

```python
prediction.setModelTypeAsDenseNet121()
prediction.setModelPath(os.path.join(execution_path, "Models/densenet121-a639ec97.pth"))
```
