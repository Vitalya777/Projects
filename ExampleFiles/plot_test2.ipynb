```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi, linspace, sin

x1 = linspace(0,2*pi,200)
y1 = sin(x1)

x2 = [0,pi/2,pi,3*pi/2,2*pi]
y2 = [0,1,0,-1,0]

# fig, ax = plt.subplots()
plt.plot(x1,y1)
plt.plot(x2,y2,marker='o')
plt.show()
```
